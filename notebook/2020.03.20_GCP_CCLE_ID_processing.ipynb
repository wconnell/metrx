```python
import glob
```

```python
def readGCP(base_dir, f_pattern):
    data_dict = {}
    files = glob.glob(os.path.join(base_dir, f_pattern))
    for f in files[:3]:
        key = os.path.basename(f).split("_")[1]
        data = pd.read_csv(f, sep='\t', index_col=0)
        meta = pd.DataFrame([row[:-1] for row in data.index.str.split("|")],
                            columns=['ENST', 'ENSG', 'OTTHUMG', 'OTTHUMT', 'GENE-NUM', 'GENE', 'NUM', 'TYPE'])
        data.index = meta['GENE']
        data_dict[key] = data.T
    return data_dict
```

```python
ccle = readGCP(base,"CCLE/CCLE_counts.tsv.gz")
ccle = ccle['counts.tsv.gz']
```

```python
# rename ids to match DepMap 'stripped cell line name'
ccle_ids = ["".join(filter(str.isalnum, i[1])).upper() for i in ccle.index.str.split(".")]
```

```python
ccle.index = ccle_ids
# drop duplicately named cell lines
ccle = ccle.reset_index().drop_duplicates('index').set_index('index')
```

```python
# DepMap CCLE metadata
ccle_dm = pd.read_csv('/srv/nas/mk2/projects/pan-cancer/DepMap/CCLE/sample_info_v2.csv', sep=",")
```

```python
# downsample to shared cell lines
ccle = ccle[ccle.index.isin(ccle_dm['stripped_cell_line_name'])]
```

```python
# downsample metadata to shared cell lines
ccle_dm = ccle_dm[ccle_dm['stripped_cell_line_name'].isin(ccle.index)]
```

```python
assert ccle.shape[0] == ccle_dm.shape[0]
```
