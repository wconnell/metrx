```python
import numpy as np
import pandas as pd

import matplotlib.pyplot as plt

# notebook settings
from IPython.core.interactiveshell import InteractiveShell
InteractiveShell.ast_node_interactivity = "all"
from IPython.display import set_matplotlib_formats
set_matplotlib_formats('retina')
```

```python
import operator as op
from functools import reduce

def ncr(n, r):
    r = min(r, n-r)
    numer = reduce(op.mul, range(n, n-r, -1), 1)
    denom = reduce(op.mul, range(1, r+1), 1)
    return numer / denom
```

```python
examples = np.arange(1, 5000)
```

```python
pairs = [ncr(n, 2) for n in examples]
```

```python
triplets = [ncr(n, 3) for n in examples]
```

```python
plt.plot(examples, examples, label='Singles')
plt.plot(examples, pairs, label='Pairs')
#plt.plot(examples, triplets)
plt.ylim((0, 20000))
plt.xlabel('Raw Samples')
plt.ylabel('Training Samples')
plt.title('Learning Task')
plt.legend()
```
