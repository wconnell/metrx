```python
import os
```

```python
root = "../data/TCGA"
sample_type = {'tumor':{}, 'normal':{}}

names=['gene', 'fpkm']
for fid,fname,ftype in zip(test['File ID'], test['File Name'], test['Sample Type']):
    if ftype=='Primary Tumor':
        sample_type['tumor'][fid] = pd.read_csv(os.path.join(root, fid, fname), sep="\t", index_col=0, header=None).T
    if ftype=='Solid Tissue Normal':
        sample_type['normal'][fid] = pd.read_csv(os.path.join(root, fid, fname), sep="\t", index_col=0, header=None).T
    print(fid, fname)
```
